```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#reading in the cleaned retail dataset
data= pd.read_excel("Retail_dataset_cleaned.xlsx")
data.head()
```

```python
#checking the shape
data.shape
```

```python
#checking the columns
data.columns
```

```python
#creating the basket
basket= data.groupby(["order_id", "product_category_name"])["order_item_id"].sum().unstack().reset_index().fillna(0).set_index("order_id")
basket.head()
```

```python
#setting everything lesser than 0 to 0 and greater than 1 to 1
def encoding_units(x):
    if x<=0:
        return 0
    else:
        return 1

basket= basket.applymap(encoding_units)
basket.head()
```

```python
# Filtering the Basket with the sum of values having >=2 to find itemsets with groups of equal to or more than 2

basket_filter = basket[(basket > 0).sum(axis=1) >=2]
basket_filter
```

```python
#now applying apriori algorithm to find frequent datasets
frequent_itemsets = apriori(basket_filter, min_support=0.05,use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)
frequent_itemsets["Length"] = frequent_itemsets['itemsets'].apply(lambda x:len(x))

frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='lift',min_threshold = 0.8).sort_values('lift', ascending=False).reset_index(drop=True)
rules
```

### Observations

**1. 97% of customers buying toys are likely to but bed_bath_table**

**2. 97% of customers buying toys are likely to buy watches_gifts**

**3. 97% of customers buying toys are likely to buy computer accessories**

**4. 97% of customers buying toys are likely to buy health_beauty**

**5. 97% of customers buying toys are likely to buy furniture_decor**

**So these are the top 5 frequent itemsets in groups of two**

```python
# Filtering the Basket with the sum of values having >=3 to find itemsets with groups of equal to or more than 3

basket_filter_2 = basket[(basket > 0).sum(axis=1) >=3]
basket_filter_2.head()
```

```python
#now applying apriori algorithm to find frequent datasets
frequent_itemsets = apriori(basket_filter_2, min_support=0.05,use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)
frequent_itemsets["Length"] = frequent_itemsets['itemsets'].apply(lambda x:len(x))

frequent_itemsets
```

```python
rules_2 = association_rules(frequent_itemsets, metric='lift',min_threshold = 0.8).sort_values('lift', ascending=False).reset_index(drop=True)
rules_2.head(20)
```
